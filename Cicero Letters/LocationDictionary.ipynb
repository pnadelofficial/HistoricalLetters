```python
f = open('C:\\Users\\brian\\Cicero Letters\\locations.csv', 'r')
LocDict = {}
for line in f: 
    placename, long, lat, notes = line.strip().split(',')
    if placename == '': 
        placename = None
    if long == '': 
        long = None
    if lat == '':
        lat = None
    if notes == '':
        notes == None
    try:
        coordinates = (float(long), float(lat))
    except:
        coordinates = (long, lat)
    LocDict[placename] = coordinates
f.close()
LocDict
```
